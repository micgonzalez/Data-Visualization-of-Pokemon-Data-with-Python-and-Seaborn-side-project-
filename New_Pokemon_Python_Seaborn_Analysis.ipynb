# New Data Visualization of Pokemon Data of Type 1 and Type 2

This notebook will outlines the use of different seaborn plots with Pokemon data. The data will be loaded into dataframes and subset in pandas. The plots will be created in seaborn and matplotlib libraries. This is an updated projected to see any new insights, can be discovered.

## Brief Overview 

I would need to import the necessary libraries to open the data with pandas. I also need to so some data wrangling to narrow down the features. After this step of the process, it will be able to see any new insights that were not seen in the last notebook.

## Importing libraries and data 

```python
# Importing the needed libraries.
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns

df = pd.read_csv('...\Pokemon.csv', index_col = 0)
df.head()
```

## Data Visualizations 

In this next section of the notebook. I will talk about my process with new insights that was discovered through seaborn plots.

## Line Plot of Type 1 Pokemon 

```python
# Line plot of all type 1 pokemon
sns.lineplot(data = df["Type 1"])
plt.title("Line Plot of Type 1 Pokemon")
plt.xlabel("Count")
```

This line plot show the distinct count of every type 1 pokemon.

## Line Plot of Type 2 Pokemon  

```python
# Line plot of all type 2 pokemon and filling in missing values for type 2 pokemon.
dt2 = df["Type 2"].fillna(axis = 0, method = 'ffill')
sns.lineplot(data = dt2)
plt.title("Line Plot of Type 2 Pokemon")
plt.xlabel("Count")
```

This line plot show the distinct count of every type 2 pokemon.

## Histogram of Defense Pokemon 

```python
# Histogram of defense type pokemon.
sns.histplot(df.Defense)
plt.title("Count of Defense Type Pokemon")
```

In this histogram it shows that there are over 35 pokemon that have defense metric of 65. This could lead to see what types of pokemon are the one with this metric.

## Bar Plot of Type 1 and Stage Defense Pokemon 

```python
# Bar plot of Type 2 defense pokemon.
sns.barplot(x = "Type 1", y = "Defense", hue = "Stage", data = df)
plt.xticks(fontsize = 5)
plt.title("Type 1 Defense Pokemon")
```

In this bar plot, it shows that for type 1 pokemon with rock and water features have higher defense values.

## Bar Plot of Type 2 and Stage Defense Pokemon 

```python
# Bar plot of Type 2 defense pokemon.
sns.barplot(x = "Type 2", y = "Defense", hue = "Stage", data = df)
plt.xticks(fontsize = 7)
plt.title("Type 2 Defense Pokemon")
```

In this bar plot, it shows that for type 2 pokemon with ground and ice features have higher defense values.

## Linear Model Plot of Pokemon's Health Points and Defense 

```python
sns.lmplot(x = "HP", y = "Defense", data = df)
```

In this linear model plot, you can see that majority of the values are in the lower health points and defense values.

## Residual Plot of Pokemon's Attack and Defense values 

```python
# Plot the residuals of the pokemon's attack and defense features.
sns.residplot(x = df.Attack, y = df.Defense, lowess = True, color = "b")
```

In this residual plot, the area near the zero line has many data points in the negative area for attack values. It seems that attack values are lower compared to the defense values.
